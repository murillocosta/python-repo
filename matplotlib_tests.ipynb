```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# gerando valores demonstrativos

x = np.arange(10)
y = np.random.normal(size=10)
```

```python
# plotar valores
plt.plot(x, y)

# definindo nomes para os eixos
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')

# defininndo um título
plt.title("Título do gráfico")

# visualizar gráfico
plt.show()
```
