```python
df = pd.read_excel(r'relatorio_de_pedidos 23DEZ22.xlsx')
df
```

```python
df_produtos_idpedido = df.loc[:, ['Produtos', 'Id do Pedido']]
df_produtos_idpedido
```

```python
## convertendo dataframe para lista
lista_pedidos_produtos = [x for x in df_produtos_idpedido['Produtos']]
lista_pedidos_produtos
```

```python
lista_pedidos_produtos_nao_tratados = []

for item_lista_pedidos_produtos in lista_pedidos_produtos:
  lista_pedidos_produtos_nao_tratados.append(item_lista_pedidos_produtos.split('\n'))

lista_pedidos_produtos_nao_tratados
```

```python
lista_pedidos_produtos_tratados = []
for lista_interna_produtos_nao_tratados in lista_pedidos_produtos_nao_tratados:
  for item_lista_interna_produtos_nao_tratados in lista_interna_produtos_nao_tratados:
    if len(item_lista_interna_produtos_nao_tratados) == 0:
      continue
    if not item_lista_interna_produtos_nao_tratados.startswith(' '):
      lista_pedidos_produtos_tratados.append(item_lista_interna_produtos_nao_tratados.rstrip())

lista_pedidos_produtos_tratados  
```

```python
matriz_produtos_quantidades = []

for item_lista_pedidos_produtos_tratados in lista_pedidos_produtos_tratados:
  matriz_produtos_quantidades.append(item_lista_pedidos_produtos_tratados.split(' x '))

matriz_produtos_quantidades
```
