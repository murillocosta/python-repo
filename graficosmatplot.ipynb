```python
# importando bibs

import pandas as pd
import matplotlib.pyplot as plt
```

# Gráfico de Linhas

```python
# importando dados para dataframe
df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/dengue-dataset.csv")
```

```python
# convertendo datas (strings) em tipo data

df['data'] = pd.to_datetime(df['data'])
```

```python
# exibindo 5 primeiros
df.head()
```

```python
plt.plot('data', 'casos-confirmados', data=df)

plt.title('Dengue em Campinas')
plt.ylabel('Casos confirmados')
plt.xlabel('Data')

plt.show()
```
